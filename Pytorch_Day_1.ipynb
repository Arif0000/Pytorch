```python
c = torch.tensor([1,-2,3,-4])
print(torch.abs(c))
print(torch.neg(c))
d = torch.tensor([1.9, 2.3, 3.7, 4.4])
print(torch.round(d))
print(torch.ceil(d))
print(torch.floor(d))
print(torch.clamp(d, min=2, max=3))
```

```python
# Reduction operation
```

```python
e = torch.randint(size=(2,3), low=0, high=10, dtype=torch.float32)
print(torch.sum(e))
print(torch.sum(e, dim=0))
print(torch.sum(e,dim=1))
print(torch.mean(e))
print(torch.mean(e,dim=1))
print(torch.median(e))
print(torch.max(e))
print(torch.min(e))
print(torch.prod(e))
print(torch.std(e))
print(torch.var(e))
print(torch.argmax(e))
print(torch.argmin(e))
```

```python
# Matrix Operation
```

```python
f = torch.randint(size=(2,3), low=0, high=10)
g = torch.randint(size=(3,2), low=0, high=10)
print(f)
print(g)
print(torch.matmul(f,g))
```

```python
# Tensor operation on GPU
```

```python
torch.cuda.is_available()
device = torch.device('cuda')

torch.rand((2,3), device=device)

a = torch.rand(2,3)
print(a)

b = a.to(device)
print(b + 5)
```

```python
import torch
import time
size = 10000

matrix_cpu1 = torch.randn(size, size)
matrix_cpu2 = torch.randn(size, size)

start_time = time.time()
result_cpu = torch.matmul(matrix_cpu1, matrix_cpu2)
cpu_time = time.time() - start_time

print(f"Time on cpu : { cpu_time:.4f} second")

matrix_gpu1 = matrix_cpu1.to('cuda')
matrix_gpu2 = matrix_cpu2.to('cuda')

start_time = time.time()
result_gpu = torch.matmul(matrix_gpu1, matrix_gpu2)
gpu_time = time.time() - start_time

print(f"Time on cpu : { gpu_time:.4f} second")

print("\nSpeedup (CPU time / GPU time):", cpu_time / gpu_time)
```
